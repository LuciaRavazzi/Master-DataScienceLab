```python
import pandas as pd
from nixtla import NixtlaClient
from dotenv import load_dotenv
import os
```

```python
dotenv_path = 'conf.env'
load_dotenv(dotenv_path)

NIXTLA_API_KEY = os.environ.get("NIXTLA_API_KEY")
NIXTLA_API_KEY
```

```python
nixtla_client = NixtlaClient(
    # defaults to os.environ.get("NIXTLA_API_KEY")
    api_key = NIXTLA_API_KEY
)
```

```python
# Read the dataset
df = pd.read_csv('https://datasets-nixtla.s3.amazonaws.com/peyton-manning.csv')

# Detect anomalies: By default, TimeGPT uses a 99% confidence interval. If a point falls outisde of that interval, it is considered to be an anomaly.
anomalies_df = nixtla_client.detect_anomalies(df, freq='D', level = 99)

# Plot anomalies
nixtla_client.plot(df, anomalies_df)
```
